```python
# signature verification
from hashlib import sha256
from ecdsa import BadSignatureError
from ecdsa.util import sigdecode_der
from ecdsa import VerifyingKey

sig_file_name = "project_signature"
message_file_name = "project_msg"
pub_key_file_name = "project_pub_key"
count = 0
matching_combo = []
matching_combo_file_name = []
"""
#step1: The below code have nested loop to check for every file with every signature file and public key file
#step2: Print the combination of message file content, signature file content and public key file content
#step3: Verification using the public key, signature and message in try block
#step4: From step3 if verification is true, then append the combination to the matching verified combinations.
#step5: After trying with all combinations, print the 10 matched combination that got from step4

"""
for i in range(1,11):
  messageFile = message_file_name + str(i) + ".txt"
  for j in range(1,11):
    sigFile = sig_file_name + str(j) + ".sig"
    for k in range(1,11):
        pubKeyFile = pub_key_file_name + str(k) + ".pem"
        combo = []
        #step 1
        with open(sigFile, "rb") as f:
            sig = f.read()
        with open(messageFile,"rb") as f:
            message = f.read()
        with open(pubKeyFile) as f:
            public_key = VerifyingKey.from_pem(f.read())

        combo.append(message)
        combo.append(sig.hex())
        combo.append(public_key.to_string().hex())
        #step 2
        print(combo)
        try:
            #step 3
            ret = public_key.verify(sig, message, sha256, sigdecode=sigdecode_der)
            assert ret
            print("Valid signature")
            count += 1
            #step 4
            if(ret is True):
              matching_combo.append(combo)
              matching_combo_file_name.append([sigFile, messageFile, pubKeyFile])
        except BadSignatureError:
            print("Incorrect signature")


#step 5
print("----------------------------------------------------------------------------------------------------------------------------------------------------------------------------------")
print("Verified Matching combinations count :", count)
print("Valid Signature and Matching combinations : \n ")
for i in matching_combo:
  print(i)
print("----------------------------------------------------------------------------------------------------------------------------------------------------------------------------------")

print("Valid signature file combinations : \n")
for i in matching_combo_file_name:
  print(i)
print("----------------------------------------------------------------------------------------------------------------------------------------------------------------------------------")

```
