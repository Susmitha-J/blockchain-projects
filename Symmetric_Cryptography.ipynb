```python
import base64
import os
from cryptography.fernet import Fernet
from cryptography.hazmat.primitives import hashes
from cryptography.hazmat.primitives.kdf.pbkdf2 import PBKDF2HMAC
import string
```

```python
password = b"1234FIM"
cipher_text = b'gAAAAABl3OPZ0r783Oo-790YGjZIYfT_8GkTl9Vz-\
9FETiZsqwo5sFzumYsqmHAw_9Ru8prHD6L_sctI8Jnu5QuUp6bqALtWaCkdcp5ndTenaY6GrY2llXDvQdMv\
doXrtYezkYdve4brZCE3lWXWv8L_uHIYujJh3w=='

password_string = password.decode('utf-8')
alphachars = list(string.ascii_uppercase)
guess_password  = None
actual_password = None
actual_salt = None
for salt_guess in range(256):  # Assuming salt is one octet
    salt = bytes([salt_guess])
    for i in alphachars:
        guess_password = None
        guess_password = password_string + i
        password = guess_password.encode('utf-8')  # Convert string back to bytes
        kdf = PBKDF2HMAC(
            algorithm=hashes.SHA256(),
            length=32,
            salt=salt,
            iterations=480000,
        )
        key = base64.urlsafe_b64encode(kdf.derive(password)) #generating key
        f = Fernet(key)
        try:
            message = f.decrypt(cipher_text) #trying to decrypt the cipher text
            print(f"\033[32mKey: {key.decode()}\033[0m")
            found = key.decode()
            actual_salt = salt
            actual_password = password
            break
        except:
            print(f"Key: {key.decode()} Salt: {salt} Password: {password}")


print("Salt:", actual_salt)
print("Guessed Password:", actual_password)
print(f"Correct Key: \033[32m{found}\033[0m\n")
print(f"Plaintext for the ciphertext is:\n\n \033[33m{message.decode()}\033[0m")


```
